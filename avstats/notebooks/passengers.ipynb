```python
# Load the Excel file using the path defined in config.py
xls = pd.ExcelFile(EXCEL_FILE)

# Check the available sheet names
print(xls.sheet_names)

# Load the specific sheets for Arrival and Departure 2023
df_arrival_2023 = pd.read_excel(xls, sheet_name='Arrival 2023', header=3)
df_departure_2023 = pd.read_excel(xls, sheet_name='Departure 2023', header=3)

# Keep only the first 11 rows for both arrival and departure
Arrival_BRU2023_StatBel = df_arrival_2023.iloc[:12].reset_index(drop=True)
Departure_BRU2023_StatBel = df_departure_2023.iloc[:12].reset_index(drop=True)

# Add an 'Arrival' column for arrival data (1 for arrival, 0 for departure)
Arrival_BRU2023_StatBel['Arrival'] = 1
Arrival_BRU2023_StatBel['Departure'] = 0

# Add a 'Departure' column for departure data (0 for arrival, 1 for departure)
Departure_BRU2023_StatBel['Arrival'] = 0
Departure_BRU2023_StatBel['Departure'] = 1

# Concatenate the two dataframes
BRU2023_StatBel = pd.concat([Arrival_BRU2023_StatBel, Departure_BRU2023_StatBel], ignore_index=True)

# Reset the index for readability
BRU2023_StatBel = BRU2023_StatBel.reset_index(drop=True)

# Display the combined dataframe
BRU2023_StatBel
```
